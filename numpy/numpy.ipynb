```python
#np.ones and np.zeros are used to create arrays of ones and zeros respectively. The shape of the array is specified as a tuple. For example, np.ones((3,4)) creates a 3x4 array of ones, and np.zeros((3,4)) creates a 3x4 array of zeros.
# np.ones((3,4),dtype=int)
np.zeros((3,4),dtype=int)
```

```python
# np.linspace explains how to create an array of evenly spaced numbers over a specified range. The function takes three arguments: the start value, the end value, and the number of samples to generate. For example, np.linspace(-100, 10, 10) generates an array of 10 evenly spaced numbers between -100 and 10.
np.linspace(-100, 10, 10)
```

```python
# np.identity creates a square identity matrix of the specified size. An identity matrix is a square matrix with ones on the main diagonal and zeros elsewhere. For example, np.identity(4) creates a 4x4 identity matrix.
np.identity(4)
```
